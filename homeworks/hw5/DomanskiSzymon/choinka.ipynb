```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def generate_christmas_tree():
    np.random.seed(67420)
    
    data = []
    layers = [(0, 5, 4), (3, 8, 3.3), (6, 10, 2.8)]

    for y_min, y_max, width in layers:
        points = 3000
        for i in range(points):
            y = np.random.uniform(y_min, y_max)
            current_width = width * (1 - (y - y_min) / (y_max - y_min))
            x = np.random.uniform(-current_width, current_width)
            data.append({'x': x, 'y': y, 'type': 'needle', 'color': 'forestgreen', 'size': 10})

    for i in range(300):
        x = np.random.uniform(-0.7, 0.7)
        y = np.random.uniform(-1.5, 0)
        data.append({'x': x, 'y': y, 'type': 'trunk', 'color': 'saddlebrown', 'size': 15})

    colors = ['red', 'gold', 'blue', 'darkorange']
    for i in range(70):
        layer = layers[np.random.randint(0, 3)]
        y_min, y_max, width = layer[0], layer[1], layer[2]
        y = np.random.uniform(y_min, y_max)
        current_width = width * (1 - (y - y_min) / (y_max - y_min))
        x = np.random.uniform(-current_width, current_width)
        data.append({'x': x, 'y': y, 'type': 'ornament', 'color': np.random.choice(colors), 'size': 80})

    data.append({'x': 0, 'y': 10.2, 'type': 'star', 'color': 'yellow', 'size': 500})

    return pd.DataFrame(data)
```

```python
df = generate_christmas_tree()

plt.figure(figsize=(8, 10), facecolor='#000814')
ax = plt.gca()
ax.set_facecolor('#000814')

for element in df['type'].unique():
    subset = df[df['type'] == element]
    plt.scatter(subset['x'], subset['y'], color = subset['color'], s = subset['size'], alpha = 0.8 if element == 'needle' else 1, edgecolors = 'none')

plt.scatter(np.random.uniform(-6, 6, 200), np.random.uniform(-2, 12, 200), color='white', s=5, alpha=0.5)

plt.axis('off')
plt.savefig('choinka.png', dpi=300, bbox_inches='tight')
plt.show()
```
